```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating the dataset
years = list(range(2010, 2021))
budget = [1457690, 1694129, 1813427, 1866586, 1988014, 2252097, 2530040, 2878283, 3132162, 3454334, 3680327]

# Creating a DataFrame
df = pd.DataFrame({'Year': years, 'Budget (in crores INR)': budget})

# Plotting the line graph
plt.figure(figsize=(10, 5))
plt.plot(df['Year'], df['Budget (in crores INR)'], marker='o', color='b')
plt.title('Annual Budget of India Over the Years')
plt.xlabel('Year')
plt.ylabel('Budget (in crores INR)')
plt.grid(True)
plt.show()

# Plotting the bar graph
plt.figure(figsize=(10, 5))
plt.bar(df['Year'], df['Budget (in crores INR)'], color='g')
plt.title('Annual Budget of India Over the Years')
plt.xlabel('Year')
plt.ylabel('Budget (in crores INR)')
plt.grid(axis='y')
plt.show()

# Plotting the scatter plot
plt.figure(figsize=(10, 5))
plt.scatter(df['Year'], df['Budget (in crores INR)'], color='r', marker='o')
plt.title('Annual Budget of India Over the Years')
plt.xlabel('Year')
plt.ylabel('Budget (in crores INR)')
plt.grid(True)
plt.show()
```
